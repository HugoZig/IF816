#Delayed Equations:

Também conhecidas como delayed equations, as equações diferenciais ordinárias com atraso são EDO's em que o seu valor depende não só do tempo e da derivada da função, mas também de um valor ocorrido em um instante de tempo passado. Ou seja, é uma EDO no seguinte formato:

  ### $\frac{\mathrm{d}x}{\mathrm{d} t} = f(x(t), x(t-\tau ), t) \mid t\geqslant t_0  $
<br>

O interessante desse tipo de ODE é sua maior semelhança ao sistema biológico humano, permitindo que, no modelo utilizado, possamos diferenciar a existência das células infectadas que podem se reproduzir e as infectadas que ainda não podem. Assim, com base nessa ideia, Baccam *et al*, formularam o seguinte modelo:


<br>

$\frac{\mathrm{d}T}{\mathrm{d} t} =  -\beta TV $

$\frac{\mathrm{d}I_1}{\mathrm{d} t} = \beta TV - kI_1$

$\frac{\mathrm{d}I_2}{\mathrm{d} t} = kI_1 - \delta I_2$

$ \frac{\mathrm{d} V}{\mathrm{d} t} = pI_2 - cV$
<br><br>
Com as seguintes variáveis:

$T$ = Número de Células não infectadas

$I_1$ = Número de Células Infectadas não reproduzentes.

$I_2$ = Número de Células Infectadas reproduzentes.

$V$ = Número de Virions

$\delta$ = Razão de Células Infectadas Mortas

$c$ = Razão das Células Infectadas Degradadas pelo Sistema Imunológico

$\beta$ = Razão pela qual Células Não Infectadas podem ser Infectadas pelas Partículas dos Vírus

$p$ = Razão em que Celulas Infectadas Produzem Partículas Virais

$k$ = Taxa de transformação de $I_1$ para $I_2$

#Uso para H1N1 influenza A

Para formular o sistema de equações acima, Baccam et al também se basearam em um conjunto de dados coletados experimentalmente acerca do vírus Influenza tipo A, responsável pela gripe H1N1. Analisando essas informações e considerando o ciclo de reprodução de um típico vírus, um modelo de delayed differential equations (DDEs) se mostrou mais eficaz em representar o avanço desses agentes no sistema imunológico, uma vez que considera uma diferença temporal entre o momento em que uma célula é infectada e o instante que ela passa a produzir viriãos (a forma infecciosa do vírus em si).

#Modelo
No artigo de Dagasso utilizado como base, pudemos entender melhor a utilização de EDO's com atraso(DDE's) e como implementá-las para a análise da resposta imunológica humana a vírus. O modelo com uso de EDO's com atraso se assemelha ao o modelo SIR. No entanto, nele, além de trabalharmos com células ao invés de individuos, temos a divisão da produção de vírus em duas categorias: células infectadas em fase de reprodução e células infectadas, mas que ainda não se reproduzem.

Coletamos artigos e publicações cientifícas acerca da reprodução viral e resposta imunológica da dengue. Com base nesses dados, iremos traçar um gráfico representativo do modelo. Veja abaixo:

```python
#Import of the required packages for model computation
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Constants used in the mathematical model are defined below
# Rate of infected cell death [cells/day]
delta = 0.5      #0.14
# Rate of degradation of infected cells by immune system [cells/day]
c = 0.5            #3.48
# Rate of which uninfected cells can be inffected [Copies/(ml * day)]
beta = 6.2* 10**-9
# Rate at which infected cells produce virus particles [Copies/(ml * day * cell)]
p = 9
# Average transition time between I1 and I2 [days]
#k = 5
k = 0.2

# X axis
t = np.linspace(0, 50, 100)

# Initial conditions
T0 = 4 * (10**8)
I1_0 = I2_0 = 0
V_0 = 10
```

```python
#Function that defines the system of ODEs to be modelled
def derivatives(y, t, delta, c, beta, p, k):
  T, I1, I2, V = y
  dTdt = - beta * T * V
  dI1dt = beta * T * V - k * I1
  dI2dt = k * I1 - delta * I2
  dVdt = p * I2 - c * V
  return dTdt, dI1dt, dI2dt, dVdt
```

```python
# Tuple that stores all the initial conditions
y0 = T0, I1_0, I2_0, V_0

# Calling Odeint with the ODE system, the initial conditions, the values for x and the constants
ret = odeint(derivatives, y0, t, args=(delta, c, beta, p, k))

# Functions calculated by Odeint
T, I1, I2, V = ret.T
```

```python
# Plot the data on four separate curves for I1(t), I2(t) T(t), V(t)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, T, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I1, 'r', alpha=0.5, lw=2, label='Infected Non−Producing Cells')
ax.plot(t, I2, 'g', alpha=0.5, lw=2, label='Infected Producing')
ax.plot(t, V, 'y', alpha=0.5, lw=2, label='Virions Number')

ax.set_xlabel('Time (days)', fontsize = 12)
ax.set_ylabel('Number of cells', fontsize = 12)
ax.set_title('Viral model', fontsize = 18)
ax.legend()
ax.legend().set_alpha(0.5)
ax.grid(which='major', c='w', lw=2, ls='-')
ax.figure.set_size_inches((12, 8))
plt.show()
```

##Resultados e Discussão:

O nosso modelo tem algumas limitações em relação às dinâmicas presentes no corpo humano ao se contrair o vírus, por exemplo, não é considerado a multiplicação e a constante renovação de células presentes no corpo ao passar do tempo, a mortalidade natural das células, ou a possibilidade delas serem degradadas pelo sistema imunológico ainda na fase I1 quando ela já é hospedeira do material genético do vírus, mas não começa ainda a ter seu organismo utilizado para produzir de fato os viriões, os vírus em sua forma infecciosa e ativa. <br>

Tendo como foco o arbovírus (vírus predominantemente disseminados por meio da picada de espécies de artrópodes) DENV, o agente responsável pela dengue, realizamos uma simulação do seu comportamento no corpo humano ao passar do tempo. O tempo de incubação, isto é, o tempo entre a infecção de uma célula e a reprodução do vírus por meio dela, do DENV pode variar de 3 a 15 dias, então, comparando os cenários entre o período mínimo (3 dias), o intermediário (9 dias) e o máximo (15) variando o parâmetro correspondente k, e analisando os gráficos feitos a partir do nosso modelo de equações diferenciais ordinárias, podemos afirmar que o comportamento do vírus é similar nesses três os casos, contudo, ainda há algumas diferenças.

Podemos observar que, ao aumentarmos o período de incubação, consequentemente diminuindo o valor de k, há um pico maior de células infectadas do tipo I1, devido ao acúmulo derivado do maior tempo de transição para passarem a produzir novos viriões, enquanto que as células do tipo I2 tendem a ter picos menores, uma vez que elas demoram mais tempo para serem originadas a partir de células I1 e a taxa de degradação de células infectadas se mantém nos três cenários, sendo assim menos provável que tenham muitas células infectadas I2 ao mesmo tempo.

Os intervalos com valores visualmente acima de zero para ambos os tipos também mudam conforme o valor de k, passando a serem mais duradouros com uma incubação mais demorada, pois, devido à maior variação de tempo para a transição entre os dois tipos, há também uma espera maior para a produção de novos viriões pelas células I2, e consequentemente, para a infecção de novas células, assim tornando a infecção do grupo inicial mais gradual e lenta.


## Conclusões

É possível notar que o problema de desenvolver um modelo matemático para um sistema tão complexo como o corpo humano, em que há uma pluralidade de variáveis e fatores que variam de maneira bastante diferente entre si, é um grande desafio. No entanto, ainda é possível criar modelos que simplificam algumas dessas variáveis e tornam o problema mais simples, permitindo a montagem de uma simulação ainda fidedigna.

Com a abordagem escolhida, foi possível observar, de início, como o tempo de incubação do vírus no corpo humano pode ser um fator decisivo para a ocorrência de casos mais sérios da dengue, reforçando as conclusões obtidas por artigos usados de referência. Essa informação pode ser utilizada para pesquisadores da área avaliarem a possibilidade de desenvolver soluções que atuem justamente nessa fase. Além disso, foi possível perceber com sucesso a forma como as diferentes componentes do sistema interagem entre si e com os parâmetros iniciais, verificando que a mudança do K gera diferentes picos e em tempos distintos.

Além disso, vale salientar que, embora os valores utilizados para o modelo tenham sido retirados de fontes confiáveis e sejam fidedignos, ainda seria pertinente validar o que foi observado. Para validar o modelo, seria interessante comparar o que foi percebido com dados experimentais, a fim de analisar sua semelhança. Uma opção seria compará-lo com uma base de dados que contenha informações sobre diferentes respostas imunológicas de organismos à dengue, como diferentes tempos de encubação, sorotipos da dengue e tempos de recuperação. Somente após essa validação, poderíamos afirmar que o modelo é capaz de representar casos reais de respostas imunológicas à dengue.

Sendo assim, verifica-se a importância da modelagem matemática munida de ferramentas e métodos computacionais para o avanço das mais distintas áreas da ciência, em particular, os tratamentos de saúde e melhoramentos no combate a doenças, possibilitando uma facilidade na formação de hipóteses e também na validação delas, pois, assim como citado por Dagasso et al, é mais barato testar *in virtu* do que em ambientes reais.
